```python
# USING Count vectoriser

import pandas as pd
import numpy as np
import ast
import spacy
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity

data = pd.read_csv("/content/Anime_data.csv")
new_data = data[['Anime_id','Title', 'Genre', 'Synopsis', 'Type', 'Producer', 'Studio', 'Source', 'Aired']]
new_data = new_data.drop(columns=['Producer','Studio'], axis=1)
new_data.dropna(axis=0, inplace=True)
new_data.reset_index(inplace=True)
new_data.drop('index', axis=1, inplace=True)

new_data.Aired.replace("Not available", np.nan, inplace=True)
new_data.dropna(axis=0, inplace=True)
new_data.Synopsis = new_data.Synopsis.apply(lambda x: x.replace(',', ''))
new_data.Aired = new_data.Aired.apply(lambda x: x.replace(',', ''))
new_data.Synopsis = new_data.Synopsis.apply(lambda x: x.split(' '))
new_data.Type = new_data.Type.apply(lambda x: x.split(' '))
new_data.Source = new_data.Source.apply(lambda x: x.split(' '))
new_data.Aired = new_data.Aired.apply(lambda x: x.split(' '))

def change():
    for i in new_data['Genre']:
        print(ast.literal_eval(i))

new_data['Genre'] = new_data['Genre'].apply(lambda x: ast.literal_eval(x))
new_data['tags'] = new_data['Genre'] + new_data['Synopsis'] + new_data['Type'] + new_data['Source'] + new_data['Aired']
new_data.drop(columns=['Genre','Synopsis','Type','Source','Aired'], axis=1, inplace=True)
new_data['tags'] = new_data['tags'].apply(lambda x: " ".join(x))

nlp = spacy.load("en_core_web_md")
def lemmatize_string(text):
    doc = nlp(text)
    lemmas = [token.lemma_ for token in doc]
    return ' '.join(lemmas)

new_data['tags'] = new_data['tags'].apply(lemmatize_string)
cv = CountVectorizer(max_features=10000, stop_words='english')
vectors = cv.fit_transform(new_data['tags']).toarray()
vectors.shape

similarity = cosine_similarity(vectors)
similarity.shape

def recommend(anime):
    index = new_data[new_data['Title'] == anime].index[0]
    similar_anime = sorted(list(enumerate(similarity[index])), reverse=True, key=lambda x: x[1])[1:11]
    return [new_data.iloc[i[0]].Title for i in similar_anime]

recommend("Bouken Ou Beet")
```

```python
# USING tfidf vectoriser

import pandas as pd 
import numpy as np 
import ast 
import spacy 

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# load data
data = pd.read_csv("/content/Anime_data.csv")
new_data = data[['Anime_id','Title', 'Genre', 'Synopsis', 'Type', 'Producer', 'Studio', 'Source', 'Aired']]
new_data = new_data.drop(columns=['Producer','Studio'],axis=1)
new_data.dropna(axis=0,inplace=True)
new_data.reset_index(inplace=True)
new_data.drop('index',axis=1,inplace=True)

# clean data
new_data.Aired.replace("Not available",np.nan,inplace=True)
new_data.dropna(axis=0,inplace=True)
new_data.Synopsis = new_data.Synopsis.apply(lambda x: x.replace(',', ''))
new_data.Aired = new_data.Aired.apply(lambda x: x.replace(',', ''))
new_data.Synopsis = new_data.Synopsis.apply(lambda x: x.split(' '))
new_data.Type = new_data.Type.apply(lambda x:x.split(' '))
new_data.Source = new_data.Source.apply(lambda x:x.split(' '))
new_data.Aired = new_data.Aired.apply(lambda x:x.split(' '))

# extract tags using genre, synopsis, type, source, and aired
new_data.Genre = new_data.Genre.apply(lambda x: ast.literal_eval(x))
new_data['tags'] = new_data['Genre'] + new_data['Synopsis'] + new_data['Type'] + new_data['Source'] + new_data['Aired']
new_data.drop(columns=['Genre','Synopsis','Type','Source','Aired'],axis=1,inplace=True)
new_data['tags'] = new_data['tags'].apply(lambda x: " ".join(x))

# lemmatize tags using spaCy
nlp = spacy.load("en_core_web_md")
def lemmatize_string(text):
    doc = nlp(text)
    lemmas = [token.lemma_ for token in doc]
    return ' '.join(lemmas)

new_data['tags'] = new_data['tags'].apply(lemmatize_string)

# convert tags to tfidf vectors
tfidf_vectorizer = TfidfVectorizer(max_features=10000, stop_words='english')
vectors = tfidf_vectorizer.fit_transform(new_data['tags']).toarray()

# calculate cosine similarity between vectors
similarity = cosine_similarity(vectors)
print(similarity.shape)

# define function to recommend similar anime
def recommend(anime):
    index = new_data[new_data['Title'] == anime].index[0]
    similar_anime = sorted(list(enumerate(similarity[index])), reverse=True, key=lambda x: x[1])[1:11]
    return [new_data.iloc[i[0]].Title for i in similar_anime]

# test recommendation system
recommend("Bouken Ou Beet")
```
